```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings


pd.set_option('display.max_columns', None)
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('/home/kotarum/Ironhack/Temario/dapt_rmt_nov23/unit 7/7.01/learningSet.csv')  
```

```python
null_values = data.isnull().sum()
null_values
```

```python
drop_list = ['OSOURCE', 'ZIP CODE']
```

```python
threshold = 0.85 * len(data)
columns_to_drop = data.columns[data.isnull().sum() > threshold]
drop_list.extend(columns_to_drop)
drop_list
```

```python

print(data['GENDER'].value_counts())
data['GENDER'] = data['GENDER'].fillna('F')

def simplify_gender(gender):
    if gender == 'M':
        return 'M'
    elif gender == 'F':
        return 'F'
    else:
        return 'Other'

data['GENDER'] = data['GENDER'].apply(simplify_gender)
print(data['GENDER'].value_counts())
```

```python
unique_genders = data['GENDER'].unique()
unique_genders
```

-------------

```python
numerical_nulls = data.select_dtypes(include=['number']).isnull().sum()
print(numerical_nulls)
```

```python
print(data['GEOCODE2'].value_counts())
data['GEOCODE2'] = data['GEOCODE2'].fillna(method='ffill')
print(data['GEOCODE2'].value_counts())
```

```python
print(data['WEALTH1'].value_counts())
data['WEALTH1'].fillna(data['WEALTH1'].median(), inplace=True)
print(data['WEALTH1'].value_counts())
```

```python
data['ADI'].fillna(data['ADI'].mode()[0], inplace=True)
data['DMA'].fillna(data['DMA'].mode()[0], inplace=True)
data['MSA'].fillna(data['MSA'].mode()[0], inplace=True)
```

```python
numerical_cols = data.select_dtypes(include=['number']).columns
data[numerical_cols].hist(figsize=(100, 100))
plt.show()
```
